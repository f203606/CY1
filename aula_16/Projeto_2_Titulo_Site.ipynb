```python
# Exibindo o título de uma página web

import requests
from bs4 import BeautifulSoup

from tkinter import *
from tkinter import messagebox

class Application:
    
    def __init__(self, master=None):
        
        self.fontePadrao = ("Arial", "14")
        
        self.primeiroContainer = Frame(master)
        self.primeiroContainer["pady"] = 10
        self.primeiroContainer.pack()

        self.segundoContainer = Frame(master)
        self.segundoContainer["padx"] = 20
        self.segundoContainer.pack()

        self.terceiroContainer = Frame(master)
        self.terceiroContainer["pady"] = 20
        self.terceiroContainer.pack()

        self.titulo = Label(self.primeiroContainer, text="Analisador de sites")
        self.titulo["font"] = ("Arial", "26", "bold")
        self.titulo.pack()

        self.senhaLabel = Label(self.segundoContainer, text="Site: ", font=self.fontePadrao)
        self.senhaLabel.pack(side=LEFT)

        self.senha = Entry(self.segundoContainer)
        self.senha["width"] = 30
        self.senha["font"] = self.fontePadrao
        self.senha.pack(side=LEFT)

        self.autenticar = Button(self.terceiroContainer)
        self.autenticar["text"] = "Verificar"
        self.autenticar["font"] = ("Calibri", "14")
        self.autenticar["width"] = 12
        self.autenticar["command"] = self.verificaSite
        self.autenticar.pack()

        self.mensagem = Label(self.terceiroContainer, text="", font=self.fontePadrao)
        self.mensagem.pack()

    #Método verificar site
    def verificaSite(self):
        try:
            site = self.senha.get()
            resultado = requests.get(site)
            resposta = resultado.text
            soup = BeautifulSoup(resposta, 'html.parser')
        except:
            messagebox.showinfo(
            title="Erro!",
            message="Favor insira um site válido")
        else:
            opcao = messagebox.askokcancel(
                    title='Analisador de sites',
                    message = f"Site: {site} \n "
                    f"Pronto para continuar?")
            self.mensagem["text"] = soup.title.string
            

root = Tk()
Application(root)
root.mainloop()
```
